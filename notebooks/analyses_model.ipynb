```python
import os
import sys
import torch

project_root = os.path.abspath("..")
sys.path.insert(0, project_root)

from gpt_lib import SentencePieceTokenizer
```

```python
model_path = "../output/mini_gpt.pth"
```

```python
checkpoint = torch.load(
        model_path,
        map_location=torch.device("cpu")
    )
```

```python
itos= checkpoint["itos"]
```

```python
tokenizer = SentencePieceTokenizer(
            max_vocab=2000,
            model_type="bpe",
            character_coverage=1.0,
        )
```

```python
with open("../data/all_cleaned_text.md", "r", encoding="utf-8") as f:
    content = f.read()
```

```python
tokenizer.build(content)
```

```python
i = tokenizer.encode("c# is a")
print(i)
```

```python
tokenizer.decode(i)
```
